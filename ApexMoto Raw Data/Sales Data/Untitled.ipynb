```python
import pandas as pd

# Read the CSV files
sales_2020 = pd.read_csv("ApexMoto Sales Data 2020.csv")
sales_2021 = pd.read_csv("ApexMoto Sales Data 2021.csv")
sales_2022 = pd.read_csv("ApexMoto Sales Data 2022.csv")

# Stack them vertically
sales = pd.concat([sales_2020, sales_2021, sales_2022], ignore_index=True)

# Save as one CSV file
sales.to_csv("Sales.csv", index=False)

print("Sales.csv created successfully!")
print(f"Total rows: {len(sales)}")
```
